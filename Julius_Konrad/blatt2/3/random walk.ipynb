```python
def random_walk(lattice_size, steps):
    n = lattice_size # lattice size
    lattice = np.zeros((n, n), dtype=np.uint8)
    r2ee = 0

    x, y = np.random.randint(0, n, 2)
    lattice[x, y] = 1
    print(x,y)

    directions = np.array([[-1 , 0], [1, 0], [0, 1], [0, -1]])
    while steps > 0:
        # draw random direction
        dx, dy = directions[np.random.randint(0, 4, 1)[0]]
        x += dx
        y += dy
        x %= n
        y %= n
        r2ee += lattice[lattice==1].shape[0]
        lattice[x, y] = 1
        steps -= 1

    print(lattice[lattice==1].shape)
    #plt.imshow(lattice)
    return r2ee
```

```python
n=np.array([200, 400, 600, 800, 1000])
r=[random_walk(500,n) for n in n]
#display(r)
plt.plot(n, r)
fit = np.polyfit(n, r, 2)
plt.plot(n, np.polyval(fit, n))
print(fit)
```
