```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as opt
import scipy as sp
```

```python
fig, ax = plt.subplots(figsize=(6,4))
k_B = sp.constants.k
N_A = sp.constants.N_A
T_C = 1450

def get__v(v, m, T):
    return (m/(2*np.pi*T*k_B))**(3/2) * 4*np.pi*v**2 * np.exp(-(m*v**2)/(2*k_B*T))

get__v_SiO2 = lambda v : get__v(v, 60.08/1000/N_A, T_C)

vs = np.linspace(-2500,2500,1000)
ax.plot(vs, get__v_SiO2(vs))

ax.set_xlabel("$v$ [m/s]")
ax.set_ylabel("$(v)$")
```
