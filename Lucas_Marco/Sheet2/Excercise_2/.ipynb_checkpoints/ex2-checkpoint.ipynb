```julia
using DelimitedFiles
using Statistics

N_200 = readdlm("analyse_200")
N_400 = readdlm("analyse_400")
N_600 = readdlm("analyse_600")
N_800 = readdlm("analyse_800")
N_1000 = readdlm("analyse_1000")

N = [200, 400, 600, 800, 1000]
R²ₑₑ = Array{Float64}(undef, length(N))
    
R²ₑₑ[1] = mean(Float64.(N_200[:,2]))
R²ₑₑ[2] = mean(Float64.(N_400[:,2]))
R²ₑₑ[3] = mean(Float64.(N_600[:,2]))
R²ₑₑ[4] = mean(Float64.(N_800[:,2]))
R²ₑₑ[5] = mean(Float64.(N_1000[:,2]));
```

```julia
using Plots

plot(N,R²ₑₑ,legend=:false)

xlabel!("N")
ylabel!("R²ₑₑ")
```

```julia
using Plots
using Statistics

ν = [log(N[i],R²ₑₑ[i])/2 for i in 1:length(N)]
println("Average ν = ", mean(ν))

plot(N,ν,legend=:false)
xlabel!("N")
ylabel!("ν")
```
